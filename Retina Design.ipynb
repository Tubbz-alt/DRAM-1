```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
import math


fig1 = plt.figure(figsize=(7, 7), dpi=90)
ax1 = fig1.add_subplot(111, aspect='equal')
x_list = []
y_list = []
for i in range(10):
    radius = 0.005 * math.sqrt(3) ** i #0.01 * i + 0.01 * i * i
    patch1 = patches.RegularPolygon(
                (0.5, 0.5),     # (x,y)
                6,              # number of vertices
                radius,            # radius
                fill=False,
                orientation=math.pi * i / 2,
            )
    ax1.add_patch(
        patch1
    )
    for v in patch1.get_verts():
        x, y = v
        x_list.append(x)
        y_list.append(y)

# fig1.savefig('reg-polygon1d.png', dpi=90, bbox_inches='tight')
plt.show()
```

```python
fig1 = plt.figure(figsize=(8, 7))
rect = [-1, -1, 1, 1]

# setting the axis limits in [left, bottom, width, height]
ax_carthesian = fig1.add_axes(rect, aspect='equal')
ax_carthesian.scatter(x_list, y_list)

ax_polar = fig1.add_axes(rect, polar=True, frameon=False, aspect='equal')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 7))

ax.scatter(x_list, y_list)
plt.show()
```
